# Description
This notebook creates the plot comparing the (log-pdf) of Normal vs. StudentT-distributions.

```python
import numpy
import pandas
import scipy.stats
from matplotlib import pyplot

import plotting
```

```python
fig, (left, right) = pyplot.subplots(dpi=140, ncols=2, figsize=(10, 5))

x = numpy.linspace(-10, 10, 300)
left.plot(x, scipy.stats.norm.pdf(x), label="Normal", color="black")
for nu in [1, 10]:
    left.plot(x, scipy.stats.t.pdf(x, df=nu), label=rf"StudentT$(\nu={nu})$")
left.legend(frameon=False, loc="upper left")
left.set(
    xlabel="x",
    ylabel="pdf(x)",
    ylim=(0, 0.52),
)

x = numpy.linspace(-30, 30, 1000)
for df in [1, 10, 50, 100]:
    right.plot(
        x,
        scipy.stats.t.logpdf(x, loc=0, scale=1, df=df),
        label=r"StudentT$(\nu=" + f"{df})" + "$",
    )
right.plot(
    x,
    scipy.stats.norm.logpdf(x, loc=0, scale=1),
    color="black",
    linestyle=":",
    label="$\mathcal{N}$",
)
right.legend(frameon=False)
right.set(
    xlabel="x",
    ylabel="log(pdf(x))",
    ylim=(-300, 0),
)
fig.tight_layout()
plotting.savefig(fig, "2.1 Normal vs. StudentT")
pyplot.show()
```
